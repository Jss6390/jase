## 4. Matplotlib
* 플롯 그래프을 그릴 때 주로 쓰이는 2D, 3D 플롯팅 패키지
* 공식 홈페이지 URL : http://matplotlib.org/

### Matplotlib 주요 함수 : import matplotlib.pyplot as plt
- plt.figure () : plot의 새로운 figure(모양)을 생성하는 함수
- plt.xlabel () : x축의 이름을 설정하는 함수 plt.ylabel() : y 축의 이름을 설정하는 함수
- plt.title () : 플롯의 제목을 설정하는 함수
- plt.xlim () : x 축의 범위를 설정하는 함수 
- plt.ylim () : y 축의 범위를 설정하는 함수
- Plt.grid () : 격자 로 표시 유무 설정
- plt.legend () : 범례 표시 함수
- plt.subplot(nrows ncols plot_number) : 서브플롯을 반환하는 함수
 - ex) subplot(211) : 2x1 개의 서브플롯으로 이루어진 플롯의 첫 번째 서브플롯
- plt.plot () : 선이나 마커를 플롯하는 함수
- plt.scatter () : x 변수 대 y 변수의 산점도를 만드는 함수
- plt.hist () : 히스토그램을 그리는 함수
- plt.boxplot (values) # 상자그림
- plt.boxplot (values, vert=0) # 횡 방향으로 바꾸기
- plt.pie (performance, labels=objects, startangle =90); plt.show () # 파이그림

### 여러 그래프 

```python
import pandas as pd
df = pd.read_excel('data/score.xlsx')

fig, axs = plt.subplots(2, 2, figsize=(15, 10)) # 2 x 2 에 해당하는 plot 들을 생성
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수') # 데이터 설정
axs[0, 0].set_title('첫 번째 그래프') # 제목
axs[0, 0].legend() # 범례
axs[0, 0].set(xlabel='이름', ylabel='점수') # x, y 축 label
axs[0, 0].set_facecolor('lightyellow') # 전면 색
axs[0, 0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학')
axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].legend()

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'])

# 네 번째 그래프
axs[1, 1].plot(df['이름'], df['사회'], color='green', alpha=0.5)

plt.show()
```

## 5. Seaborn
* Matplotlib 를 기반으로 하는 Python 시각화 패키지
* 통계적 그래픽을 제공하는 고차원의 인터페이스 제공
* 공식 홈페이지 URL : https://seaborn.pydata.org

### Seaborn 주요 함수 : import seaborn as sns
- sns.kdeplot(data) # 확률 밀도함수
- sns.histplot(data, bins, color, kde =) # density plot + histogram( kde=False : density plot 제외)
- sns.barplot(x, y, data,color)
- sns.countplot(x, data) / sns.countplot (x, df)
- sns.boxplot (x, y, data, color) # orient='h' : 가로로 표시
- sns.violinplot(x, y, data, color)
- sns.jointplot(x, y, data, color) # scatterplot+barplot
- sns.pairplot(data, hue, palette) # 대각선 kde plot, 각 그래프는 scatter plot(결측치 제거 필요)
- sns.lmplot(x, y, data) / sns.regplot (x, y, data)
- sns.heatmap(df.corr(), annot =True, fmt ='.2f', cmap="RdBu_r") # annot=True : map 에 숫자 기재

```python
import seaborn as sns
```

```python
# Density plot 그리기
sns.kdeplot(df['키']) # kdeplot : 커널함수를 겹침으로써 히스토그램보다 부더러운 형태로 표현
plt.show() 
```

```python
# histplot으로 histogram 함께 그리기
sns.histplot(df['키'], bins=7, color='skyblue', kde=True) # bins ; 막대개수, 기본 10, kde=True : 커널함수 보여줌
plt.show()
```

```python
# 다중 histplot
키_북산 = df.query('학교== "북산고"')['키']
키_능남 = df.query('학교== "능남고"')['키']

sns.histplot(키_북산, bins=3, color='blue', kde=False) # kde=False : density plot 생략
sns.histplot(키_능남, bins=3, color='gray', kde=False)
```

```python
# barplot
sns.barplot(x='학교', y= '국어', data=df) # y의 평균값으로 자동계산해줌
```

```python
# countplot
sns.countplot(x='학교', data=df)
```

```python
# boxplot
sns.boxplot(x='학교', y='키', data=df)
```

```python
# violinplot
sns.violinplot(x='학교', y='키', data=df, color='gray')
```

```python
# scatter plot
sns.scatterplot(x='국어', y='영어', hue='학교', data=df)
```

```python
# joint plot
sns.jointplot(x='국어', y='영어', data=df)
```

```python
# Pairplot
sns.pairplot(df, vars=['국어', '영어', '수학'], hue='학교')
```

```python
# heatmap
sns.heatmap(df.corr(), annot=True, fmt='.2f', linewidths=0.5, cmap="RdBu_r") # annot=True : map에 숫자 기재
```

```python
# lm plot
sns.lmplot(x='국어', y='영어', data=df) # sns.regplot(x='국어', y='영어', data=df)과 동일
```
